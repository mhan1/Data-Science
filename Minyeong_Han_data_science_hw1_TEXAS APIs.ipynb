### Complete the following tasks.

1. Find all of the restaurants listed with a zip code of `75002` and store that list of dictionaries in the variable labled `results`.

```python
url11 = "https://data.texas.gov/resource/naix-2893.json?location_zip=75002"
```

```python
import requests
response11 = requests.get(url11)
results = response11.json()
```

```python
results
```

```python
len(results)
# 1000
```

Now make another API call, and display only those results who had total receipts of 0.

```python
url22 = "https://data.texas.gov/resource/naix-2893.json?location_zip=75002&$where=total_receipts=0"

response22 = requests.get(url22)
```

```python
results_with_zero = response22.json()
```

```python
results_with_zero
```

```python
results_with_zero[98]
```

```python
results_with_zero[98]['location_zip']
```

```python
results_with_zero[98]['total_receipts']
```

```python
len(results_with_zero)
# 98 
```

Next let's try to find the top five results in that zip code with the highest total_receipts.  

> Hint: Think about how this is done in SQL.

```python
url33 = "https://data.texas.gov/resource/naix-2893.json?location_zip=75002&$order=total_receipts DESC&$limit=5"

response33 = requests.get(url33)
```

```python
results_of_highest = response33.json()
```

```python
results_of_highest
```

```python
len(results_of_highest)
```

```python
list(map(lambda result: result['total_receipts'], results_of_highest))
# ['403845', '359448', '344838', '341086', '333538']
```
