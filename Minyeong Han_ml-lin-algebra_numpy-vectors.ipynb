Display the three amounts where beer sales most outpace liquor sales.

```python
import numpy as np

diff1 =  beer_receipts - liquor_receipts

beer_more_than_liquor = np.sort(diff1)[::-1] #reversing the sort
print(beer_more_than_liquor)

beer_more_than_liquor[0:3]
                     
# array([341, 317, 310])
```

```python
liquor_more_than_beer = beer_more_than_liquor[::-1] #reversing the array
print(liquor_more_than_beer)

liquor_more_than_beer[0:3]

# array([-439, -418, -354])
```
