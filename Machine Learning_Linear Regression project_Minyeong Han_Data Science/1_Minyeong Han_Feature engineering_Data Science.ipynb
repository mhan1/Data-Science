### My created method 7: finding unique values of multiple columns.

```python
# My created method 7: finding unique values of multiple columns.

import numpy as np

def finding_unique(df, column_list):
    if not isinstance(column_list, list):
        column_list = [column_list]
    unique_value = []
    for column in column_list:
        unique_values = df[column].unique()
        column_name = column
        unique_value.append(column_name)
        unique_value.append(unique_values)
        uniques = np.array(unique_value)
    return np.array([unique for unique in uniques])
```

### My created method 8: Finding booleans

```python
#finding booleans that has only 2 values to convert to booleans by mapping.
# I modified the method "find_booleans" into simpler way as below.
# My modified method : finding booleans

import numpy as np

def find_booleans(df):   
    columns = df.columns  
    boolean_columns = np.array([column for column in columns if len(df[column].value_counts()) == 2])
    boolean_values = np.array([df[column].unique() for column in boolean_columns])
    columns_and_values = np.stack((boolean_columns, boolean_values)).T
    return columns_and_values
```

### My created method 9: I modified 'summarize_counts(df)' method to return the top values in descending order of frequencies for each 'object' feature that I have.

```python
# My created method : Returning the top values in descending order of frequencies for each 'object' feature
    
def summarize_counts(df):
    object_features = find_object_features(df)
    frequencies = np.array([df[column].value_counts(normalize=True).values[0] for column in object_features]).reshape(-1, 1)
    columns = np.array(object_features).reshape(-1, 1)
    top_values = np.array([df[column].value_counts(normalize=True).index[0] for column in object_features]).reshape(-1, 1)
    summarize = np.hstack((columns, frequencies, top_values)) 
    return summarize[summarize[:, 1].argsort()[::-1]] 
```

### My created method 10: finding the almost binary features to convert into boolean.

```python
# My created method : finding the almost binary features that has threshold of 0.80 frequency.

def almost_binary(df, threshold = .80 ):
    return np.array([np.array([cat, top]) for cat, freq, top in summarize_counts(df) if 1.0 > float(freq) > threshold])
```

```python
def remove_punctuation(string):
    return string.strip().lower().replace(' ', '_').replace('(', '').replace(')', '').replace(',', '')
```

```python
def matrix_new_features(df):
    bin_feats = almost_binary(df)
    new_bin_feats = np.array(['{column}_is_{top}'.format(column = column, top = remove_punctuation(top)) for column, top in bin_feats])
    return np.hstack((bin_feats[:, 0].reshape(-1, 1), bin_feats[:, 1].reshape(-1, 1), new_bin_feats.reshape(-1, 1)))
```

```python
matrix_new_features(df1)
```

```python
def booleans_without_top_values(df, not_values_list):
    potential_new_features = matrix_new_features(df)
    not_tf = ~np.isin(potential_new_features[:, 1], not_values_list)
    return potential_new_features[not_tf]
```

```python
booleans_without_top_values(df1, ['t', 'f'])
```

```python
def almost_to_boolean(df):
    columns_to_replace = matrix_new_features(df)[:, 0]
    values_to_replace = matrix_new_features(df)[:, 1]
    new_column_names = matrix_new_features(df)[:, 2]
    to_replace_df = pd.DataFrame({})
    for column, value, new_name in zip(columns_to_replace, values_to_replace, new_column_names):
        bool_column = np.where(df[column] == value, 1, 0)
        to_replace_df[new_name] = bool_column
    return to_replace_df
```

```python
def selected_cat_values(column, threshold = .05):
    values_counted = column.value_counts(normalize=True)
    return values_counted[values_counted > threshold]
```

```python
def reduce_cat_values(column, threshold = .05):
    column = column.copy()
    selected_values = selected_cat_values(column, threshold).index
    column[~column.isin(selected_values)] = 'other'
    column.astype('category', inplace=True)
    return column
```

```python
def df_reduced_categories(df, categoricals, threshold = .05):
    new_df = pd.DataFrame()
    for category in categoricals:
        new_df[category] = reduce_cat_values(df[category], threshold)
    return new_df
```
