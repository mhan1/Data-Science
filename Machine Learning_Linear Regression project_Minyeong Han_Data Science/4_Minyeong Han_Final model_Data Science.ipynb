### Pulling my data with selected scaled features

### Start by retrieving my data with selected scaled features in the previous step.

```python
import pandas as pd

selected_scaled_df = pd.read_csv('https://raw.githubusercontent.com/mhan1/Data-Science/master/Machine%20Learning_Linear%20Regression%20project_Minyeong%20Han_Data%20Science/pruned_scaled_features.csv')
selected_scaled_df.head(3)
```

```python
#removing unnecessary column 
selected_scaled_df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
selected_scaled_df.head().T
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

selected_scaled_X = selected_scaled_df.iloc[:, :-1]
selected_scaled_y = selected_scaled_df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(selected_scaled_X, selected_scaled_y, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

model1 = LinearRegression()
model1.fit(X_train, y_train)
model1.coef_
```

```python
model1.score(X_test, y_test)
```

```python
model1.score(X_val, y_val)
```

```python
import numpy as np

def feature_importances(df, estimator, df_cols = None):
    df_cols = df_cols or df.columns
    df_cols = df_cols.to_numpy()
    coefs = estimator.coef_
    abs_coefs = abs(coefs)
    paired = np.hstack((df_cols.reshape(-1, 1), coefs.reshape(-1, 1), abs_coefs.reshape(-1, 1)))
    ordered_cols = paired[paired[:, -1].argsort()][::-1]
    return ordered_cols
```

```python
feature_importances(selected_scaled_X, model1)
```

### model1 was trained with scaled_X and scaled_y after feature selection, and model2 was trained with unscaled_X and unscaled_y after the feature selection.<br>The score against the respective validation set was the same as 0.2828 for both model1 and model2. The score against the respective test set was also the same as 0.1956 for both model1 and model2.

### However, the coefficients for model1 and model2 were different, as shown above "feature_importance" output for each model.<br>For the model1 trained with scaled_X and scaled_y, 'fuel_type_is_gas' was the most important feature having the coefficient of -0.82 to predict the target variable "normalized_losses".<br> For the model2 trained with unscaled_X and unscaled_y, 'fuel_type_is_gas' is the most important feature but the coefficient was -87.33.<br> The other coefficients for each feature in respective model were different from each other, this is because the unit of each feature in the model2 is different from each other, while model1 has the same unit(z-score) for each feature. Hence, when it comes to talking about the feature importance, we should use scaled features like model1. When the unit is normalized, such as z-score, we can say how much our target variable change given an expected amount of movement in the feature.

### Through my analsis of the automobile data, I identified a set of core features that drive the normalized_loss values.<br>The target variable, normalized_loss represents the average loss per car per year. It is the relative average loss payment per insured vehicle year, and this value is normalized for all autos within a particular size classfication (two-door small, station wagons, sports/speciality, etc.)<br>According to the model1 feature importance,'fuel_type_is_gas', 'compression_ratio' were the top two features that change the normalized_loss value, followed by 'engine_size', 'symboling', 'drive_wheels_rwd', and so on.<br>The least important feature was 'engine_location_is_front', 'highway_mpg', 'make_volkswagen', 'curb_weight',  'bore', and so on.<br>I could see that the normalized losses in use as compared to other cars depend on symboling which was initially assigned risk factor to cars, with higher value if it is more risky. The negative cofficients for the top two important features such as 'fuel_type_is_gas' and 'compression_ratio' indicate that using gas as fuel type instead of diesel, and higher compression ratio (the ratio of the maximum to minimum volume in the cylinder of an internal combustion engine) contribute to lowering the normalized losses. A higher compression ratio means that the engine is getting more work out of the same amount of fuel, which is good for power and also miles per gallon. Higher compression ratio also means more thermal efficiency. Hence, it was good to confirm the negative coefficient for the 'compression_ratio' feature, as I could expect that higher thermal efficiency or miles per gallon could reduce the normalized losses.<br>In fact, 'horsepower' and 'highway_mpg' were also can be the measure for the thermal efficiency, and their respective coefficients appeared to be negative, as epxected. Going forward, I could reduce the features by eliminating these two 'horsepower' and 'highway_mpg' as 'compression_ratio' can represent these features and it is the factor that most contribute to normalized_loss among these three features.<br>Also, 'drive_wheels_rwd' and 'drive_wheels_fwd' appeared to drive normalized_loss, but with different sign. 'drive_wheels_rwd' increases the normalized_loss, while 'drive_wheels_fwd' decreases the normalized_loss based on the sign of respective coefficient.  A drive wheel is a wheel of a motor vehicle that transmits force, and front-wheel drive (FWD) vehicles' engines drive the front wheels. Using the front wheels for delivery of power as well as steering allows the driving force to act in the same direction as the wheel is pointing. Based on the different sign of coefficient for 'drive_wheels_rwd' and 'drive_wheels_fwd' features, I think this results can also help the car manufacturers to determine which type of drive wheel they should make.<br>Interestingly, several 'make' of cars appeared to contribute to the normalized_loss, with different coefficients, respectively. 'make_peugot' and 'make_nissan' have respective positive coefficients, which means these 'make' contribute to normalized_loss. Meanwhile, 'make_honda', 'make_volvo', 'make_volkswagen' have respective negative coefficients, meaning these 'make' contribute to lowering the normalized_loss. If I work for the insurance company, I would prefer 'honda', 'volvo', 'volkswagen' to other make, to reduce the normalized_loss value.<br>It was very interesting to see which feature drive the normalized_loss by using linear regression algorithm for this machine learning project.
